```python
def avg(v: list[float]) -> float:
    return sum(v) / len(v)
```

**jSON file-ok beolvasása**

```python
import json
sztr_bsc = {}
sztr_prnt = {}
sztr_all = {}
with open('eredmenyek(ErtekAdas).json', encoding='utf-8') as file:
    sztr_bsc = json.load(file)

with open('eredmenyek(print).json', encoding='utf-8') as file:
    sztr_prnt = json.load(file)

with open('eredmenyek(osszes).json', encoding='utf-8') as file:
    sztr_all = json.load(file)

v_bsc = sztr_bsc['Futasi eredmenyek']
v_prnt = sztr_prnt['Futasi eredmenyek']
v_all = sztr_all['Futasi eredmenyek']
```

*Átlag eredmények (if, ter, notif)*

```python
ifis_basic: list[float] = [s['Elágazással'] for s in v_bsc]
ter_basic: list[float] = [s['Ternary-val'] for s in v_bsc]
notif_basic: list[float] = [s['Elágazás nélkül'] for s in v_bsc]

avgif_basic = float(f"{avg(ifis_basic):.5f}")
avter_basic = float(f"{avg(ter_basic):.5f}")
avnot_basic = float(f"{avg(notif_basic):.5f}")

ifis_print: list[float] = [s['Elágazással'] for s in v_prnt]
ter_print: list[float] = [s['Ternary-val'] for s in v_prnt]
notif_print: list[float] = [s['Elágazás nélkül'] for s in v_prnt]

avgif_print = float(f"{avg(ifis_print):.5f}")
avter_print = float(f"{avg(ter_print):.5f}")
avnot_print = float(f"{avg(notif_print):.5f}")

ifis_all: list[float] = [s['Elágazással'] for s in v_all]
ter_all: list[float] = [s['Ternary-val'] for s in v_all]
notif_all: list[float] = [s['Elágazás nélkül'] for s in v_all]

avgif_all = float(f"{avg(ifis_all):.5f}")
avter_all = float(f"{avg(ter_all):.5f}")
avnot_all = float(f"{avg(notif_all):.5f}")

with open('stat.txt', 'w', encoding='utf-8') as fl:
    print("SEBESSÉGMÉRÉS ÉRTÉKADÁSNÁL:\n-----------------------------", file=fl)
    print(f"{avgif_basic=}\n{avter_basic=}\n{avnot_basic=}", file=fl)
    print("\nSEBESSÉGMÉRÉS PRINT UTASÍTÁSSAL:\n-----------------------------", file=fl)
    print(f"{avgif_print=}\n{avter_print=}\n{avnot_print=}", file=fl)
    print("\nSEBESSÉGMÉRÉS MINDKETTŐVEL:\n-----------------------------", file=fl)
    print(f"{avgif_all=}\n{avter_all=}\n{avnot_all=}", file=fl)
```
